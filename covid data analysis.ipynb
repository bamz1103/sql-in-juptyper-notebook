#### Looking at total cases vs total deaths 

```python
death_percentage = '''
SELECT location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 AS death_percentage
FROM covid_deaths_analysis
ORDER BY 1,2
'''
pd.read_sql(death_percentage,engine)
```

#### Looking at the total cases vs population showing what percentage of population got covid

```python
infected_percentage = '''
SELECT location, date, total_cases, population, (total_cases/population)*100 AS infected_percentage
FROM covid_deaths_analysis
WHERE continent IS NOT NULL
ORDER BY 1,2
'''
pd.read_sql(infected_percentage,engine)
```

#### Looking at countries with highest infection rate compared to population

```python
maximum_percentage_infected = '''
SELECT location, MAX(total_cases) AS highest_infection_count, population, MAX((total_cases/population))*100 AS maximum_percentage_infected
FROM covid_deaths_analysis 
WHERE continent IS NOT NULL
GROUP BY location, population
ORDER BY maximum_percentage_infected DESC
LIMIT 1000
'''
pd.read_sql(maximum_percentage_infected,engine)

#percentage_infected.to_excel("percentageinfected.xlsx", index = False)
```

#### showing countries highest death count per population

```python
total_death_count = '''
SELECT location, MAX(total_deaths) as total_death_count
FROM covid_deaths_analysis 
WHERE continent IS NOT NULL
GROUP BY location
ORDER BY total_death_count DESC
'''
pd.read_sql(total_death_count,engine)
```

#### showing continents with highest death count per population

```python
total_death_count_continent = '''
SELECT location, SUM(new_deaths) AS total_death_count
FROM covid_deaths_analysis 
WHERE continent IS NULL
AND location NOT IN ('World', 'High income', 'Upper middle income', 'Lower middle income', 'European Union','Low income')
GROUP BY location
ORDER BY total_death_count DESC
'''
pd.read_sql(total_death_count_continent,engine)
#continent_death_count1.to_excel("continentdeathcount.xlsx", index=False)
```

#### Global death percentage per day 

```python
global_death_percentage_perday = '''
SELECT date, SUM(new_cases) as total_cases, SUM(new_deaths) AS total_deaths, SUM(new_deaths)/NULLIF(SUM(new_cases),0)*100 AS Global_death_percentage
FROM covid_deaths_analysis 
WHERE continent IS NOT NULL
GROUP BY date
'''
pd.read_sql(global_death_percentage_perday,engine)
```

#### Global death percentage till date(04/10/2023)

```python
global_death_percentage = '''
SELECT SUM(new_cases) as total_cases, SUM(new_deaths) AS total_deaths, SUM(new_deaths)/NULLIF(SUM(new_cases),0)*100 AS Global_death_percentage
FROM covid_deaths_analysis 
WHERE continent IS NOT NULL
'''
global_death = pd.read_sql(global_death_percentage,engine)
global_death.to_excel("globaldeath.xlsx", index=False)
```

#### Viewing the covid vaccination analysis tabble looking at the total population vs vaccinations

```python
population_vs_vaccinations = '''
SELECT dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations 
FROM covid_deaths_analysis AS dea
JOIN covid_vaccinations_analysis AS vac ON dea.location = vac.location AND dea.date = vac.date
WHERE dea.continent IS NOT NULL
ORDER BY 2,3
'''
pd.read_sql(population_vs_vaccinations,engine)
```

#### Looking at the rolling count of new_vaccinations to get the accurate total number of new_vaccinations for each day where the vaccinations were administered 

```python
total_new_vaccinations_perday = '''
WITH PopvsVac (Continent, Location, Date, Population, New_Vaccination, RollingPeopleVaccinated) AS
(
SELECT dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations, SUM(vac.new_vaccinations) 
OVER (PARTITION BY dea.location ORDER BY dea.location, dea.date) AS RollingPeopleVaccinated 
FROM covid_deaths_analysis AS dea
JOIN covid_vaccinations_analysis AS vac ON dea.location = vac.location AND dea.date = vac.date
WHERE dea.continent IS NOT NULL
)
SELECT *, (RollingPeopleVaccinated/population)*100 FROM PopvsVac
ORDER BY 1,2
'''
pd.read_sql(total_new_vaccinations_perday,engine)
```

```python
percentpopulationinfected = '''
SELECT location, population, date, MAX(total_cases) AS highestinfectioncount, MAX(total_cases/population)*100 AS percentpopulationinfected
FROM covid_deaths_analysis
WHERE location NOT IN ('World', 'High income', 'Upper middle income', 'Lower middle income', 'European Union','Low income')
GROUP BY location, population, date
ORDER BY percentpopulationinfected DESC
'''
percentpopulationinfect = pd.read_sql(percentpopulationinfected,engine)
percentpopulationinfect.to_excel("globaldeat.xlsx", index=False)
```
